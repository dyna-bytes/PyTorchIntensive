```python
# 변수의 shape, type, value 확인
def p(var,_str='') :
    if _str=='\n' or _str=='cr' :
        _str = '\n'
    else :
        print(f'[{_str}]:')
        _str = ''
    if type(var)!=type([]):
        try:
            print(f'Shape:{var.shape}')
        except :
            pass
    print(f'Type: {type(var)}')
    print(f'Values: {var}'+_str)

def pst(_x,_name=""):
    print(f'[{_name}] Shape{_x.shape}, {type(_x)}')
def ps(_x,_name=""):
    print(f'[{_name}] Shape{_x.shape}')
```

#### **n x n ploting function**

```python
# plot n x n images
def square(imgs, n):
    img_num = n
    plt.figure(figsize=(6,6))
    for i in range(n):
        for j in range(n):
            p_num = i*n + j
            # specify subplot and turn of axis
            ax = plt.subplot(n, n, p_num+1)   # pic position num: 1 ~ 64
            ax.set_xticks([])
            ax.set_yticks([])
            # plot filter channel in grayscale
            plt.imshow(imgs[0, p_num, :, :], cmap='gray')
    # show the figure
    plt.show()
```

## **첫번째 conv layer의 feature map 시각화**

```python
# plot n x n images
def square(imgs, n):
    img_num = n
    plt.figure(figsize=(6,6))
    print(f'min[{np.min(imgs):2.2f}],max[{np.max(imgs):2.2f}]')
    f_min, f_max = imgs.min(), imgs.max() #(3,3,64,64)
    imgs = (imgs - f_min) / (f_max - f_min)
    for i in range(n):
        for j in range(n):
            p_num = i*n + j
            ax = plt.subplot(n, n, p_num+1)#(row,col,index)
            ax.set_xticks([])
            ax.set_yticks([])
            # plot filter channel in grayscale
            plt.imshow(imgs[p_num, :, :], cmap='gray')
    plt.show()
```

## **VGG모델의 모든 conv layer Feature map 시각화 하기**


```python
class VGGFeatureMap(nn.Module):
    def __init__(self, features):
        super(VGGFeatureMap, self).__init__()
        self.features = features
    def forward(self, x):
        out1 = self.features[:3](x)
        out2 = self.features[:8](x)
        out3 = self.features[:15](x)
        out4 = self.features[:22](x)
        out5 = self.features[:29](x)
        return out1, out2, out3, out4, out5

vggFeatureMap = VGGFeatureMap(model_vgg.features)
#summary(vggFeatureMap, input_size=(8, 3, 224, 224))
```

```python
## 모델에 image넣어 실행
feature_maps = vggFeatureMap(input_batch.to(device))
# plot the output from each block
for i, fmap in enumerate(feature_maps):
    print(f'[{i+1} Block] Features :',fmap[0].shape)
    square(fmap[0].cpu().detach().numpy(), 8)
```
